```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the CSVm data set
dataset=pd.read_csv("movies.csv")

#plotting scatter plot of total_bill vs tip
plt.scatter(dataset['movie'],dataset['genre'])

#giving our plot a title
plt.title("this is scatter plot")

#giving x and y labels names
plt.xlabel('movie')
plt.ylabel('genre')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the CSVm data set
dataset=pd.read_csv("movies.csv")

#plotting scatter plot of total_bill vs tip
plt.bar(dataset['movie'],dataset['genre'])

#giving our plot a title
plt.title("Bar chart")

#giving x and y labels names
plt.xlabel('movie')
plt.ylabel('genre')
plt.show()
```

```python
#import the required package
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#reading the csv data set using pandas
dataset=pd.read_csv("movies.csv")
sns.lineplot(x='movie',y='genre',data=dataset)
plt.show()
```

```python
#importing the required modules
import bokeh.plotting
import pandas as pd

#creating an object for the figure
chart=bokeh.plotting.figure(title="Bokeh Bar Chart")

#reading the csv dataset through pandas
dataset=pd.read_csv("movies 2.csv")

#counting unique columns of the tips file
count=dataset['genre'].value_counts()

#plotting the graph
chart.line(count,dataset['genre'])

#showing the figure
bokeh.plotting.show(chart)
```

```python
#importing the required modules
import pandas as pd
import plotly.express

#reading the csv dataset through pandas
dataset=pd.read_csv("movies.csv")

#plotting our scatter plot
graph=plotly.express.scatter(dataset,x="movie",y="genre")

#displaying the plot created
graph.show()

```
