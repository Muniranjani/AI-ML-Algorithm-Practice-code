```python
import pandas as pd
df=pd.read_csv('cybersecurity_attacks.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
##filling the null values
```

```python
df2=df.fillna(value=0)
df2
```

```python
df2.isnull().sum().sum()
```

```python
#filling the null values with previous value

df3=df.fillna(method='pad')
df3
```

```python
#filling null values with next value

df4=df.fillna(method='bfill')
df4
```

```python
#filling null values with backward values

df5=df.fillna(method='pad',axis=1)
df5
```

```python
#filling null values with forward values

df7=df.fillna(method='bfill',axis=1)
df7
```

```python
#filling different values in null in different columns

df8=df.fillna({'Proxy Information':'234.5.895.456',
              'Firewall Logs':'log data'})
df8
```

## dropna()

```python
df10=df.dropna()
df10
```

```python
df11=df.dropna(how='any')
df11
```

```python
df12=df.dropna(axis=1)
df12
```

## replace()

```python
import numpy as np
df13=df.replace(to_replace=np.nan,value=20)
df13
```

```python
df14=df.replace(to_replace=17616,value=27616)
df14
```

## interpolate()

```python
df['Proxy Information']=df['Proxy Information'].interpolate(method='linear')
df
```

```python
df['Proxy Information']=df['Proxy Information'].interpolate(method='linear',limit_direction='forward')
df
```
